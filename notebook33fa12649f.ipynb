```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
data =pd.read_csv('../input/housedata/data.csv')
X=data.iloc[:100,1]
Y=data.iloc[:100,4]
theta0=-1
theta1=-1
alpha=0.01
print(Y)
```

```python
meanX=statistics.mean(X)
meanY=statistics.mean(Y)
stdX=statistics.stdev(X)
stdY=statistics.stdev(Y)
for i in range(0,len(X)):
    X[i]=(X[i]-meanX)/stdX
    Y[i]=(Y[i]-meanY)/stdY
```

Plotting X vs Y

```python
plt.scatter(X,Y)
plt.show()
```

```python
def h_x(theta0,theta1,x):
    return theta0+(theta1*x)
def gradient(theta0,theta1,x,y):
    temp0=theta0
    temp1=theta1
    sum0=0
    sum1=0
    m=len(x)
    k=0
    while(k<20):
        sum0=0
        sum1=0
        for i in range(0,len(x)-25):
            sum0=h_x(theta0,theta1,x[i])-y[i]
            sum1=(h_x(theta0,theta1,x[i])-y[i])*x[i]
        temp0=temp0-(alpha*(sum0)/m)
        temp1=temp1-(alpha*(sum1)/m)
        temp0=round(temp0)
        temp1=round(temp1)
        if(temp0==theta0 and temp1==theta1):
            return int(theta0),int(theta1)
        theta0=round(temp0)
        theta1=round(temp1)
        k=k+1
    return int(theta0),int(theta1)
        
```

```python
a=gradient(theta0,theta1,X,Y)
print(a)
```
